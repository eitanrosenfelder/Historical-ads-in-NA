```python
def pic(link_p):
    link=[]
    page = requests.get(link_p,headers = headers)
    the_tree = html.fromstring(page.content)
    tite = the_tree.xpath('//div[@class="document col-xs-12 col-sm-6 col-md-4 blacklight-image"]/div/a')
    for i in range(len(tite)):
        x = str(etree.tostring(tite[i])).split(">")
        y=x[0].split('="')
        link.append(y[2])
    next_p=the_tree.xpath('//a[@rel="next"]')
    z = str(etree.tostring(next_p[0])).split(">")
    w=z[0].split('="')
    v=w[2][:-1]
    if v== 'return false;" hre':
            next_link= []
    else:
        strin="https://repository.duke.edu"
        next_link=str(strin+v)
    return(link,next_link)
```

```python
def collecting(link):
    strin="https://repository.duke.edu"
    for j in link:
        sleep(random.randint(6, 8))
        add=str(strin+j)[:-1]
        site=requests.get(add,headers = headers)
        tree = html.fromstring(site.content)
        title = tree.xpath('//dd[@class="blacklight-title_ssi"]/ul/li//text()')
        head=tree.xpath('//dd[@class="blacklight-headline_tesim"]/ul/li//text()')
        time= tree.xpath('//dd[@class="blacklight-date_tesim"]/ul/li//text()')
        company=tree.xpath('//dd[@class="blacklight-company_tesim"]/ul/li/a//text()')
        product=tree.xpath('//dd[@class="blacklight-product_tesim"]/ul/li/a//text()')
        category=tree.xpath('//dd[@class="blacklight-category_tesim"]/ul/li/a//text()')
        subject=tree.xpath('//dd[@class="blacklight-subject_tesim"]/ul/li//text()')
        pub=tree.xpath('//dd[@class="blacklight-publication_tesim"]/ul/li//text()')
        medium=tree.xpath('//dd[@class="blacklight-medium_tesim"]/ul/li//text()')
        formaty=tree.xpath('//dd[@class="blacklight-format_tesim"]/ul/li//text()')
        permalink=tree.xpath('//dd[@class="blacklight-permanent_url_ssi"]/a//text()')
        line.append(head)
        name.append(title)
        date.append(time)
        comp.append(company)
        prod.append(product)
        cate.append(category)
        subj.append(subject)
        publication.append(pub)
        medi.append(medium)
        forma.append(formaty)
        perma.append(permalink)
```
