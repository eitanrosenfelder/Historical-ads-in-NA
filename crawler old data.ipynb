```python
from bs4 import BeautifulSoup
import requests
import requests.exceptions
from urllib.parse import urlsplit
from urllib.parse import urlparse
from collections import deque
from time import sleep
import random
from lxml import html
from lxml import etree
import os
import pandas as pd
```

```python
counter = 0
headers = {
    'User-Agent': 'Amitai Gispan and Eitan Rosenfelder, Hebrew university of Jerusalem Israel',
    'From': 'eitan.rosenfelder@mail.huji.ac.il'}


page = requests.get("https://repository.duke.edu/dc/adaccess?f%5Bcommon_model_name_ssi%5D%5B%5D=Item",headers = headers)
the_tree = html.fromstring(page.content)
tite = the_tree.xpath('//div[@class="document col-xs-12 col-sm-6 col-md-4 blacklight-image"]/div/a')
```

```python
link=[]
name=[]
date=[]
comp=[]
prod=[]
cate=[]
subj=[]
publication=[]
medi=[]
forma=[]
perma=[]
```

```python
strin="https://repository.duke.edu"
for i in range(len(tite)):
    x = str(etree.tostring(tite[i])).split(">")
    y=x[0].split('="')
    link.append(y[2])
```

```python
def collecting(link):
    for j in link:
    #https://repository.duke.edu
        sleep(random.randint(8, 15))
        add=str(strin+j)[:-1]
        site=requests.get(add,headers = headers)
        tree = html.fromstring(site.content)
        title = tree.xpath('//dd[@class="blacklight-title_tesim"]/ul/li//text()')
        time= tree.xpath('//dd[@class="blacklight-date_tesim"]/ul/li//text()')
        company=tree.xpath('//dd[@class="blacklight-company_tesim"]/ul/li/a//text()')
        product=tree.xpath('//dd[@class="blacklight-product_tesim"]/ul/li/a//text()')
        category=tree.xpath('//dd[@class="blacklight-category_tesim"]/ul/li/a//text()')
        subject=tree.xpath('//dd[@class="blacklight-subject_tesim"]/ul/li//text()')
        pub=tree.xpath('//dd[@class="blacklight-publication_tesim"]/ul/li//text()')
        medium=tree.xpath('//dd[@class="blacklight-medium_tesim"]/ul/li//text()')
        formaty=tree.xpath('//dd[@class="blacklight-format_tesim"]/ul/li//text()')
        permalink=tree.xpath('//dd[@class="blacklight-permanent_url_ssi"]/a//text()')
        name.append(title)
        date.append(time)
        comp.append(company)
        prod.append(product)
        cate.append(category)
        subj.append(subject)
        publication.append(pub)
        medi.append(medium)
        forma.append(formaty)
        perma.append(permalink)
    return name, date, comp, prod, cate, subj, publication, medi, forma, perma
```

```python
def convert(name,date, comp, prod, cate, subj, publication, medi, forma, perma):
    df = pd.DataFrame(name)
    df["year"] = date
    df["company"] = comp
    df["product"] = prod
    df["category"] = cate
    df["subject"] = subj
    df["publication"] = publication
    df["medium"] = medi
    df["format"] = forma
    df["link"] = perma
    return df
```
